```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. CHARGEMENT (Depuis le Mart validé précédemment)
df = pd.read_gbq("SELECT * FROM `ton_projet.dataset_marts.mart_analysis_full`")

# Configuration du style pour un rendu "pro"
sns.set_theme(style="whitegrid")

# 2. ANALYSE UNIVARIÉE : Distribution des prix
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], kde=True, color='skyblue')
plt.title('Distribution Globale des Prix')
plt.xlabel('Prix (MAD)')
plt.show()

# 3. ANALYSE BIVARIÉE : Boxplots par plateforme
# Cela montre la dispersion et les prix médians par retailer
plt.figure(figsize=(12, 6))
sns.boxplot(x='source', y='price', data=df, palette='viridis')
plt.title('Comparaison de la Dispersion des Prix par Plateforme')
plt.xlabel('Plateforme')
plt.ylabel('Prix (MAD)')
plt.show()

# 4. ANALYSE TEMPORELLE : Évolution des prix
# Si tu n'as pas la date dans le mart_analysis_full,
# tu peux utiliser une jointure avec la table cleaned
query_time = """
    SELECT DATE(scraped_at) as date, source, AVG(price) as avg_price
    FROM `ton_projet.dataset.prices_cleaned`
    GROUP BY 1, 2
"""
df_time = pd.read_gbq(query_time)

plt.figure(figsize=(14, 6))
sns.lineplot(data=df_time, x='date', y='avg_price', hue='source', marker='o')
plt.title('Évolution Temporelle du Prix Moyen par Plateforme')
plt.xticks(rotation=45)
plt.show()
```
