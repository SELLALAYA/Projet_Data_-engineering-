```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# 1. CHARGEMENT (Tu utilises les colonnes is_amazon, is_jumia que nous avons créées)
df = pd.read_gbq("SELECT * FROM `ton_projet.dataset_marts.mart_analysis_full`")

# 2. MATRICE DE CORRÉLATION (Heatmap)
# On vérifie si discount_pct ou rating impactent le prix
plt.figure(figsize=(8, 6))
corr = df[['price', 'discount_pct', 'rating']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matrice de Corrélation : Facteurs influençant le Prix')
plt.show()

# 3. RÉGRESSION LINÉAIRE MULTIPLE (OLS)
# Formule : price = β0 + β1*discount + β2*rating + β3*is_amazon + β4*is_jumia
X = df[['discount_pct', 'rating', 'is_amazon', 'is_jumia']]
X = sm.add_constant(X)  # Ajout de la constante (Intercept)
y = df['price']

model = sm.OLS(y, X).fit()

# Affichage des résultats
print(model.summary())

# 4. EXTRACTION DES COEFFICIENTS POUR INTERPRÉTATION
print("\n--- Analyse des Coefficients ---")
coeffs = model.params
for feature, coef in coeffs.items():
    print(f"L'impact de {feature} sur le prix est de : {coef:.2f} MAD")
```
